```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
start_date = "2023-01-01"
end_date = pd.Timestamp.today().strftime('%Y-%m-%d') 
ticker = "BTC-USD"
df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=True)
df.head()
```

```python
df.tail()
```

```python
df = df.reset_index()
df.head()
```

```python
plt.plot(df.Close)
```

```python
# moving average (ma)
ma100 = df.Close.rolling(100).mean()
ma200 = df.Close.rolling(200).mean()
```

```python
plt.figure(figsize = (12,6)) # size of graph
plt.plot(df.Close) # plot close on graph
plt.plot(ma100,'r') # plot data after 100 days as green line 
plt.plot(ma200,'y') # plot data after 200 days as yellow line
```
